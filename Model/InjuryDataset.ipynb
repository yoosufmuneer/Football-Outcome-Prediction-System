```python
import pandas as pd
import numpy as np
```

```python
#df1 = pd.read_csv('Final-player.txt',error_bad_lines=False, header=None)
#df1.head()
df = pd.read_csv('PlayerInjuries.json',error_bad_lines=False, sep=',', header=None)
df.head()

cols_to_check = [2,10,12,13,14]

#Remove Unnecessary Characters
df[cols_to_check] = df[cols_to_check].replace({'u':''},regex=True)
df[cols_to_check] = df[cols_to_check].replace({'n':''},regex=True)
df[cols_to_check] = df[cols_to_check].replace({"\'":''},regex=True)

df = df.dropna(subset=[14,15])

#months = ['Aug','Sep','Oct','Nov','Dec','Jan','Feb','Mar','Apr','May']

mask = df[14].str.contains('18/19')
dfEighteen = df[mask]

#August Data
august = df[15].str.contains('Aug')
dfEighteenAug = dfEighteen[august]

#September Data
september = df[15].str.contains('Sep')
dfEighteenSep = dfEighteen[september]

#October
october = df[15].str.contains('Oct')
dfEighteenOct = dfEighteen[october]

#November
november = df[15].str.contains('Nov')
dfEighteenNov = dfEighteen[november]

#December
december = df[15].str.contains('Dec')
dfEighteenDec = dfEighteen[december]

#January
january = df[15].str.contains('Jan')
dfEighteenJan = dfEighteen[january]

#February
february = df[15].str.contains('Feb')
dfEighteenFeb = dfEighteen[february]

#March
march = df[15].str.contains('Mar')
dfEighteenMar = dfEighteen[march]

#April
april = df[15].str.contains('Apr')
dfEighteenApr = dfEighteen[april]

#May
may = df[15].str.contains('May')
dfEighteenMay = dfEighteen[may]
dfEighteenMay.head()

#Combine All Monthly Dataframes
injuryDF2018 = pd.concat([dfEighteenAug, dfEighteenSep, dfEighteenOct, dfEighteenNov,dfEighteenDec,dfEighteenJan,
                   dfEighteenFeb,dfEighteenMar,dfEighteenApr,dfEighteenMay])
#format Date
day = injuryDF2018[15].str.split(" ",expand=True)[2]
month = injuryDF2018[15].str.split(" ",expand=True)[1];
#Remove Unneccsary Characters
month = month.replace({"u'":''},regex=True)
injuryDF2018[16] = injuryDF2018[16].replace({' ':'/'},regex=True)
injuryDF2018[16] = injuryDF2018[16].replace({"'":''},regex=True)
injuryDF2018[16] = injuryDF2018[16].replace({"]":''},regex=True)
injuryDF2018[16] = injuryDF2018[16].replace({"\)":''},regex=True)

#Map the number of the month to the date in the injury data
d = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Aug':'08', 'Sep':'09', 'Oct':'10','Nov':'11','Dec':'12'}
month = month.map(d)

#Create new data column with the matching date format to stats Dataset
injuryDF2018["FullDate"] = day+"/"+month+injuryDF2018[16]
injuryDF2018['FullDate'] = injuryDF2018['FullDate'].replace({"u'":''},regex=True)

#Remove Unnecessary Characters
injuryDF2018[1] = injuryDF2018[1].replace({"u'":''},regex=True)

#Correct Misspelled Team Names
injuryDF2018[2] = injuryDF2018[2].replace({" Flham FC":'Fulham'},regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({" Watford FC":'Watford'},regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({" Newcastle Uited":'Newcastle'},regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({" Hddersfield Tow":'Huddersfield'},regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({" Liverpool FC":'Liverpool'},regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Chelsea FC':'Chelsea'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Arseal FC':'Arsenal'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Leicester City':'Leicester'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Crystal Palace':'Crystal Palace'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Sothampto FC':'Southampton'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Cardiff City':'Cardiff'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' AFC Boremoth':'Bournemouth'}, regex=True)
injuryDF2018[2] = injuryDF2018[2].replace({' Wolverhampto Waderers':'Wolves'}, regex=True)

#injuryDF2018 = injuryDF2018[injuryDF2018[2].str.contains('"\\"')]
injuryDF2018 = injuryDF2018[[1,2,10,14,'FullDate']]
FulhamDS = injuryDF2018[injuryDF2018[2]=="Fulham"]
WatfordDS = injuryDF2018[injuryDF2018[2]=="Watford"]
NewcastleDS = injuryDF2018[injuryDF2018[2]=="Newcastle"]
HuddersfieldDS = injuryDF2018[injuryDF2018[2]=="Huddersfield"]
LiverpoolDS = injuryDF2018[injuryDF2018[2]=="Liverpool"]
ChelseaDS = injuryDF2018[injuryDF2018[2]=="Chelsea"]
ArsenalDS = injuryDF2018[injuryDF2018[2]=="Arsenal"]
LeicesterDS = injuryDF2018[injuryDF2018[2]=="Leicester"]
CrystalPalaceDS = injuryDF2018[injuryDF2018[2]=="Crystal Palace"]
SouthamptonDS = injuryDF2018[injuryDF2018[2]=="Southampton"]
CardiffDS = injuryDF2018[injuryDF2018[2]=="Cardiff"]
BournemouthDS = injuryDF2018[injuryDF2018[2]=="Bournemouth"]
WolvesDS = injuryDF2018[injuryDF2018[2]=="Wolves"]


FinalDF =  pd.concat([FulhamDS, WatfordDS, NewcastleDS, HuddersfieldDS,LiverpoolDS,ChelseaDS,
                   ArsenalDS,LeicesterDS,CrystalPalaceDS,SouthamptonDS,CardiffDS,BournemouthDS,WolvesDS])


FinalDF[1] = FinalDF[1].replace({"'":''},regex=True)
namesList = ['Sergio Rico', 'Domingos Quina']

for i in namesList:
  indexNames = FinalDF[ FinalDF[1].str.contains(i)].index
  FinalDF.drop(indexNames , inplace=True)

FinalDF[10] = FinalDF[10].replace({"Defeder":'Defender'},regex=True)
FinalDF[10] = FinalDF[10].replace({"Defesive":'Defendsive'},regex=True)
FinalDF[10] = FinalDF[10].replace({"Cetral":'Central'},regex=True)
FinalDF[10] = FinalDF[10].replace({"Attackig":'Attacking'},regex=True)
FinalDF[10] = FinalDF[10].replace({"Wiger":'Winger'},regex=True)
FinalDF[10] = FinalDF[10].replace({"Cetre":'Centre'},regex=True)

#new_col = ["15/08/2018", "05/12/2018", "25/08/2018","11/02/2019","31/05/2019","31/05/2019","19/04/2019","01/03/2020","11/09/2019","31/05/2019","31/05/2019"."31/05/2019","31/05/2019","22/04/2019","31/05/2019","31/05/2019","19/05/2019","26/04/2019","31/05/2019","31/05/2019"] 
```
